```
#Data points to plot
years = [1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009]
letters = [9,8,11,12,11,13,12,9,9,7,9]
deaths = [ 6,5,5,10,8,14,10,4,8,5,6]

#creating a plot
fig = plt.figure( figsize=(15,5))
#adding a subplot to the figure
ax = fig.add_subplot(111)
#to have grid on the plot
ax.grid()

#plotting spelling bee letters for each decade
lns1 = ax.plot(years, letters, color='maroon', marker="D",label = 'Spelling Bee Winning Word')
#creating a twin x-axis to make it common for plotting deaths caused by spiders
ax2 = ax.twinx()
#plotting 2nd graph with same x axis for deaths by spider in each decade
lns2 = ax2.plot(years, deaths, color='black', marker="o",label = 'Number of people killed by venomous spider')
#setting label for left Y axis
ax.set_ylabel(r"Spelling Bee Winning Word",labelpad=10,fontsize='large', color= 'maroon')
ax.set_xticks(years)#setting ticks for common x axis
#6,6,3,2
#3,3,1,0

#fig text is used to add tittle with multiple lines and different style
plt.figtext(0.2,1,"Letters in Winning Word of Scripps National Be \n\n\n", color='maroon',ma='center',fontsize=25, family='serif',weight='bold')
plt.figtext(0.4,1,"correlates with \n\n\n", color='grey',ma='center',fontsize=20, family='serif')
plt.figtext(0.2,1," Number of People Killed by Venemous Spider \n", color='black',ma='center',fontsize=25,family='serif',weight='bold')
plt.figtext(0.4,1,"correlation:80.57% (r=0.8057)", color='grey',ma='center',fontsize=15, family='serif')

#tick params to enable plot to copy x-axis in top and bottom
ax.tick_params(reset=True,which="major",axis="x",bottom=True, top=True, labeltop=True)
ax.set_yticks([5,10,15])#setting ticks of letters plot y axis 
ax.set_yticklabels(labels=['5 lettrs','10 letters', '15 letters'], color= 'maroon')# setting labels for letters on y axis

#setting label for paralell y-axis on right side of deaths
ax2.set_ylabel(r"Number of people killed by venomous spider",labelpad=20, fontsize=15,rotation=270)
ax2.set_yticks([0,5,10,15]) # setting y axis ticks of right side to plot deaths
plt.ticklabel_format(useOffset=False) # to disable predefined settings of tick labels
#setting labels for second y axis to show deaths by spiders
ax2.set_yticklabels(labels=["0 deaths","5 deaths","10 deaths","15 deaths"])

plt.tight_layout()
# adding legends of both lines plotted for letters and deaths
ax.legend(loc= 'upper center',bbox_to_anchor=(0.7, -0.05), ncol=2, borderaxespad = 2,fontsize="large")
ax2.legend(loc= 'upper center',bbox_to_anchor=(0.4, -0.05), ncol=2, borderaxespad = 2,fontsize="large",framealpha = 10)
plt.show()
```

**Part 3**  
**Reproducing graphs of overlapping data**

```
#using pandas read_csv function to read data directly instead of accessing local data
vehicles = pd.read_csv("https://raw.githubusercontent.com/pratoolbharti/NIU/master/Applied_Machine_Learning_S20/Homework-1/mpg.csv")

#data frame is created as df of vehicles
df = pd.DataFrame(vehicles)
#figure and subplots is created with 4 rows and 1 column grid figure
fig, ax = plt.subplots(4, 1, figsize=(8,28))

#random function to generate random value for jitter which returs new random value for each point but close value
def rand_jitter(temp):
  std_dev = temp.max()/100
  return temp + np.random.randn(len(temp)) * std_dev

#random function to generate random value for jitter which returs new random value for each point but not close value
def rand_jitter1(temp):
  std_dev = temp.max()/50
  return temp + np.random.randn(len(temp)) * std_dev
#jitter function with right amount of transperancy to make data understandale through visualization
def jitter(x, y, c):
  return ax[2].scatter(rand_jitter(x), y, s=100, marker='o',label=c, alpha=0.7,edgecolor='black',linewidth=0.8,cmap=None)

#jitter function with different values to show how bad the data could be visualized if over jittering
def jitter1(x, y, c):
  return ax[3].scatter(rand_jitter1(x), y, s=100, marker='o',label=c, alpha=0.6,edgecolor='black',linewidth=0.6)

# creating a loop through group of types of vehicle engine to plot different engine with unique color
# here c is to get the type of engine and df can be used as dataframe for that group
for c, df in df.groupby('drv'):
  # plotting data of displacement and city on scatter plot
  ax[0].scatter(df['displ'], df['cty'], s=100,label=c)
  # same data is plotted with right amount of transperancy
  ax[1].scatter(df['displ'], df['cty'], s=100,label=c, alpha=0.6,edgecolor='black',linewidth=0.5)
  # same data can also be visualized more accurately through jitter with new values but close to originals
  jitter(df['displ'],df['cty'],c)
  # same data is plotted but this function plots data to show what could go wrong if jittered is applied more
  jitter1(df['displ'],df['cty'],c)
  

#disabling spines of all subplots from top and right
ax[0].spines['right'].set_visible(False)
ax[0].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)
ax[1].spines['top'].set_visible(False)
ax[2].spines['right'].set_visible(False)
ax[2].spines['top'].set_visible(False)
ax[3].spines['right'].set_visible(False)
ax[3].spines['top'].set_visible(False)

#setting x and y labels for 1 st subplot and y ticks
ax[0].set_xlabel("displacement (|)", fontsize=15)
ax[0].set_ylabel("fuel economy (mpg)", fontsize=15)
ax[0].set_yticks((10,15,20,25,30,35))

#setting x and y labels for 2nd subplot and y ticks
ax[1].set_xlabel("displacement (|)", fontsize=15)
ax[1].set_ylabel("fuel economy (mpg)", fontsize=15)
ax[1].set_yticks((10,15,20,25,30,35))

#setting x and y labels for 3rd subplot and y ticks
ax[2].set_xlabel("displacement (|)", fontsize=15)
ax[2].set_ylabel("fuel economy (mpg)", fontsize=15)
ax[2].set_yticks((10,20,30))

#setting x and y labels for 4th subplot and y ticks
ax[3].set_xlabel("displacement (|)", fontsize=15)
ax[3].set_ylabel("fuel economy (mpg)", fontsize=15)
ax[3].set_yticks((10,20,30))

#legends are created for each subplot with labels
ax[0].legend(title='drive train',labels=('4WD','FWD','RWD'))
ax[1].legend(title='drive train',labels=('4WD','FWD','RWD'))
ax[2].legend(title='drive train',labels=('4WD','FWD','RWD'))
ax[3].legend(title='drive train',labels=('4WD','FWD','RWD'))
plt.show()
```
